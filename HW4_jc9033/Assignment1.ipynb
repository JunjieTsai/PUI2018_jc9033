# Enviroment preparation

```python
import numpy as np
import pylab as pl
%pylab inline
```

```python
import matplotlib.mlab as mlab
```

```python
from scipy.stats import norm
```

```python
mean = 10
np.random.seed(1994)
def SizeN():
    return np.random.randint(10,2001)
```

# 1. Normal Distribution

```python
# Generate Dataset

Normal = {}
Normal_Means = {}

for i in range(100):
    Normal[i] = np.random.normal(loc = mean, size = SizeN())
    Normal_Means[i] = Normal[i].mean()
```

```python
# Scatter Plot

ax_Normal = pl.figure(figsize=(5,5)).add_subplot(111)
for i in Normal:
    ax_Normal.plot(Normal[i].size , Normal_Means[i] , 'o')
    ax_Normal.plot([0,2000], [mean, mean], 'k')
    ax_Normal.set_xlabel('Sample Size' , fontsize = 14)
    ax_Normal.set_ylabel('Sample Mean' , fontsize = 14)
    ax_Normal.set_title('Normal Distribution' , fontsize = 18)
```

## Figure 1a: In Normal Distribution, the larger the sample size is, the closer the sample mean is to the presupposed mean, which coincided with the Central Limit Theorem.

```python
# Histogram Plot

Normal_Means = list(Normal_Means.values())

pl.figure(figsize=(5, 5))
pl.hist(Normal_Means,bins=30)
pl.xlabel('Sample Mean', fontsize = 14)
pl.ylabel('Sample Number', fontsize = 14)
pl.title('Normal Distribution', fontsize = 18)
```

## Figure 1b: In Normal Distribution, the larger the sample size is, the closer the sample mean is to the presupposed mean. According to the Central Limit Theorem, the histogram approximates a normal distribution.

# Extra Credit: Gaussian fit for the histogram

```python
# gaussian
fig = pl.figure(figsize=(5, 5))

ax = fig.add_subplot(111)
ax.hist(All_Means,bins=30,normed=True)

xmin, xmax = pl.xlim()
x = np.linspace(xmin,xmax,100)
#y = norm.pdf(x,np.mean(All_Means),np.std(All_Means))
y = mlab.normpdf(x , np.mean(All_Means) , np.std(All_Means))
ax.plot(x,y,'red',label=u'Gaussian fit curve')

pl.xlabel('Sample Mean', fontsize = 14)
pl.ylabel('Sample Number', fontsize = 14)
pl.title('All Distributions', fontsize = 18)
pl.legend()
```

## Figure 7: In the integration of all 5 above distributions, it can be well-fitted with a Gaussian Fit Curve, which means the histogram approximates a normal distribution. This observation coincided with the Central Limit Theorem.
